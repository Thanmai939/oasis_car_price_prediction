```python
from sklearn.preprocessing import LabelEncoder
a=['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
```

```python
le=LabelEncoder()
for i in a:
    df[i]=le.fit_transform(df[i])
    
```

```python
df
```

```python
X=df.drop('price',axis=1)
X
```

```python
Y=df['price']
Y
```

```python
#splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

# Model Training


```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python

predict=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,r2_score

error=r2_score(Y_train,predict)
print("R squared error: ",error)
```

```python
plt.scatter(Y_train,predict)
plt.xlabel('Actual price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Prices')
plt.show()
```

```python
test_predict=model.predict(X_test)
```

```python

error=r2_score(Y_test,test_predict)
error
```

```python
plt.scatter(Y_test,test_predict)
plt.xlabel('Actual price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Prices')
plt.show()
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso()
lasso.fit(X_train,Y_train)
```

# MOdel evaluation

```python
train_predict=lasso.predict(X_train)
```

```python
error=r2_score(Y_train,train_predict)
print('R squared error: ',error)
```

# visualize the actual prices and predicted prices

```python
plt.scatter(Y_train,train_predict)
plt.xlabel('Actual price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Prices')
plt.show()
```
